```python
from sklearn.datasets import load_iris

import pandas as pd
import numpy as np

np.random.seed(0)
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)

#Just adding the Y column, classification column
df['Species Class'] = pd.DataFrame(iris.target)

df
```

```python
X = df
X = X.drop(['Species Class'], axis = 1)

Y = pd.DataFrame()
Y['Species Class'] = df['Species Class']

print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(X, Y, train_size = 0.7, stratify = Y, 
                                                 shuffle = None, random_state = np.random.randint(low = 1, high = 50))

# (shuffle = True/False) will shuffle or nor the train and test data
# (stratify = y/None) will make sure sure the test & train set it proportionate and not
# --------Cannot set shuffle and stratify both at a time---------------------

#random_state deals with the randomness of the split operation
#It's customary to add both train_size & test_size. It will automatically calculate from any one of
#                         the train_size and test size(Which will be given)
#stratify = y --> will make sure the test and train data is well proportionate
#to shuffle and disregard the proportion, shuffle = True can be set in the place of stratify
```

```python
print("Number of training examples: ", len(train_x))
print("Number of testing examples: ", len(test_x))
```

```python
train_x.head()
```

```python
train_y.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_jobs = 4, random_state = 0)
clf.fit(train_x, train_y.values.ravel())
```

```python
test_x.head()
```

```python
test_data = test_x
test_data.head()
```

```python
def name(sl):
    if (sl == '0'): return 'setosa'
    elif(sl == '1'): return 'versicolor'
    else: return 'virginica'
```

```python
prediction = clf.predict(test_x)
P = pd.DataFrame({'Prediction': prediction[:]}, index = test_data.index)
P.head()

test_data = test_data.join(test_y)
test_data = test_data.join(P)

test_data
```

```python
#Naming the classes

test_data['Species Class'] = test_data['Species Class'].astype(str)
test_data['Prediction'] = test_data['Prediction'].astype(str)

for i in range(len(test_data)):
    pos = test_data.index[i]
    test_data.loc[pos, 'Species Class'] = name(test_data['Species Class'][pos])
    test_data.loc[pos, 'Prediction'] = name(test_data['Prediction'][pos])

test_data
```

```python
y_test = test_data['Species Class']
predicted = test_data['Prediction']
```

```python
CM = pd.crosstab(y_test, predicted)
```

```python
from sklearn.metrics import multilabel_confusion_matrix

CM_multiclass = multilabel_confusion_matrix(test_data['Species Class'], test_data['Prediction'], 
                            labels = ["setosa", "versicolor", "virginica"])
```

```python
from sklearn.metrics import precision_recall_fscore_support as score
precision, recall, fscore, support = score(y_test, predicted)

print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))
```
